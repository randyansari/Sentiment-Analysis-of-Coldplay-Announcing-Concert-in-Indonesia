```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
# Tentukan kueri pencarian
query = '"coldplay" near:Indonesia since:2023-05-09 until:2023-05-15'
```

```python
# Ambil tweet dan simpan dalam sebuah list
tweets = []
for tweet in sntwitter.TwitterSearchScraper(query).get_items():
    tweets.append([tweet.date, tweet.rawContent])
```

```python
# Konversi list ke DataFrame dan simpan sebagai file CSV
df = pd.DataFrame(tweets, columns=['Date', 'Tweets'])
df.to_csv('coldplay_tweets.csv', index=False)
```
